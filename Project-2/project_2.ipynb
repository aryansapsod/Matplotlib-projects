```python
import matplotlib.pyplot as plt
import pandas as pd
#load data
df = pd.read_csv('amazon.csv')
#clean Data
df = df.dropna(subset=['category','discounted_price','actual_price'])
product_count=df['actual_price'].value_counts()
plt.figure(figsize=(8,6))
plt.bar(product_count.index,product_count.values,color=['skyblue','orange'])
plt.title('product count vs values')
plt.xlabel('product counts')
plt.ylabel('product values')
plt.savefig('counts_values.png')
# plt.xticks('')
category_1 =df['category'].value_counts()

plt.figure(figsize=(8,6))
plt.plot(category_1.index,category_1.values,color=('purple'))
plt.title('Category Counts By Plotting')
plt.xlabel('No.of Count')
plt.ylabel('Category Name')
# plt.tight_layout()
plt.savefig('category_1.png')
plt.show()

discounted_price_1 =df['discounted_price'].value_counts()
# actual_price =df['actural_price'].value_counts
plt.barh(discounted_price_1.index ,discounted_price_1.values,color=('orange'))
# plt.pie(actual_price.index)


```
